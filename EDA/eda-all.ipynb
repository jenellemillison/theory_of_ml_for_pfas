## Data Ingestion

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# loop to create dict of dataframes, each df in dfs is one data table

# adjust file path as needed
folder_path = r'C:\Users\andre\OneDrive - Johns Hopkins\Documents\GitHub\theory_of_ml_for_pfas\data\raw_data'
dfs = {}

for f in os.listdir(folder_path):
    # only include data files, not metadata
    if f.endswith('.csv') and f != 'PFAS_Data_Dictionary.csv':
        key_name = os.path.splitext(f)[0]
        fp = os.path.join(folder_path, f)
        dfs[key_name] = pd.read_csv(fp,encoding='ISO-8859-1')
```
